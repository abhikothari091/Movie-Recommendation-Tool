```python
# we need to extract the names from these list of dictionaries
#we need to create a custom fuction

def convert(obj):
    list =[]
    for i in obj:
        list.append(i['name'])
    return list
```

```python
# as we can see its a string so it is not able to parce through properly
# we nned to convert this string of list to a list
#for that we need a module
import ast 
```

```python
def convert(obj):
    list =[]
    for i in ast.literal_eval(obj):
        list.append(i['name'])
    return list
```

```python
movies['genres'].apply(convert)
# now we get a proper list
```

```python
def convert3(obj):
    list =[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!=3:
            list.append(i['name'])
            counter+=1
        else:
            break
    return list
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
# now for crew
movies['crew'][0]
# here we need only the director so the job value should be director
```

```python
#now we create another function to extract the directors and stop the loop when we find a director
def fetch_director(obj):
    list =[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            list.append(i['name'])
            break
    return list
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
#now here, to recoomend movies, we have to use the movie tags which the user enters
# further we have to use the tags of that movie and find 5 movies with similarity tags
# for this we need vectorization: convert this text into vectors
# so there will be 5000 vectors for 5000 movies , thus we have to find the closest vector to the user entered movie to recoomned
# this techniques is called text vectorization : "bag of words"
# what we do is we combine all the movie tags for a movie then we use this large text to find most common words amongst this
#that is calculate the 5000 words of top frequency, so suppose a word actionn appeards 5 times, future appears 3 times and thus we create an entire table for frequesncy pf each word
# now we compare this with the table of another movie for all these words , the shape of table will be (5000,5000) i.e words and movies
#  ow we plot these vectors based on the frequency and then find the closest vector to the correspondingvector of user input movie
# we can also use more than 5000 words but it is best to find the optimum number of words
# we wont consider stop words i.e words use for sentence formation eg. and,or,to, from, etc, they dont ad meaning to the sentence
# and pewrform vectorization on other words
# we have scikit learn which has a class called count vectorizer
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
# here we create an object cv and here we add functions like max_feature to determine number of vectors we want to use
# also add a function stop words in order to remove english language stop words
```

```python
cv.fit_transform(df['tags']).toarray()
#it would have a lot of values as zero and we convert this to a numpy array
```

```python
# now we need nltk library to do stemming
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#now we create a helper function
def stem(text):
    y = []
    
    for i in text.split():
        ps.stem(i)
```

```python
ps.stem('loving')
```

```python
ps.stem('loved')
```

```python
ps.stem('dancing')
```

```python
#as we can see it gives the stem word
def stem1(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
# as we can see many words are altered to their stem words
df['tags'] = df['tags'].apply(stem1)
```

```python
# now the words are not repeating
# and every movie is converted into vectors and we have 4800 movies and that many vectors as well and 5000 numbers in each one
#now we have to find the distance between a movie with every other movie, we will calculate cosine distance and not euclidien distance
# we will find the angle between them (cosine distance) if its the same movie then the angle will be 0
# so the larger the angle, the larger the dissimilarity amongst the movie
# euclidien dist is not reliable for comparison
# also distance is inversely proportional to similarity
# we have a function in cosine similarity in sci kit learn
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
#now we will get the distance of each vector with every other
```
